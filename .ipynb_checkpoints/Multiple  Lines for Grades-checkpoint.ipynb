```python
from django.shortcuts import render
from student.models import Grade

#sql to pandas
import pandas
from django.db import connection

#bokeh
from bokeh.charts import Scatter, TimeSeries
from datetime import datetime
from collections import OrderedDict
from bokeh.plotting import figure

#for notebook only
from bokeh.charts import show
from bokeh.io import output_notebook
```

```python
grades_sql = "SELECT grade, grade_date, subject_id  FROM student_grade  \
           WHERE student_id = '%s'" %('50155809')  
df_grades = pandas.read_sql(grades_sql, con=connection)


grades_sql = "SELECT grade, grade_date, subject_name  FROM student_grade, student_subject  \
           WHERE student_grade.student_id = '%s' AND student_grade.subject_id=student_subject.subject_id" %('50155809', )
df_grades2 = pandas.read_sql(grades_sql, con=connection)
df_grades
```

```python
df_math = df_grades[df_grades.subject_id==1]
df_ela = df_grades[df_grades.subject_id==2]
df_ss = df_grades[df_grades.subject_id==3]
df_sci = df_grades[df_grades.subject_id==4]
```

```python
grades_dict2=df_grades.set_index('grade_date').to_dict('index')
grades_dict2
```

```python
grades_dict = OrderedDict({'grade_date': df_grades.grade_date})
grades_dict['grade'] = df_grades.grade
grades_dict['subj_id'] = df_grades.subject_id
grades_dict
 
```

```python
# create a new plot with a datetime axis type
#legacy chart, uses dictionary
p3 = figure(width=800, height=250, x_axis_type="datetime")
p3.line(grades_dict['grade_date'], grades_dict['grade'], color='navy', alpha=0.5)
output_notebook()
show(p3)
```

```python
from bokeh.models import  TapTool, HoverTool, ColumnDataSource, DatetimeTickFormatter
source = ColumnDataSource(data=grades_dict)

TOOLS = ['hover', 'pan', 'tap']

plot = figure(x_axis_type='datetime', plot_height=250, tools=TOOLS)

plot.line('grade_date', 'grade', line_width=5, source=source)


output_notebook()
show(plot)
```

```python
from math import pi
from datetime import datetime as dt

from bokeh.models import DatetimeTickFormatter


df = pandas.DataFrame(data=[1,2,3],
                  index=[dt(2015, 1, 1), dt(2015, 1, 2), dt(2015, 1, 3)],
                  columns=['foo'])
p2 = figure(plot_width=400, plot_height=400)
p2.line(df.index, df['foo'])
p2.xaxis.formatter=DatetimeTickFormatter(formats=dict(
        hours=["%d %B %Y"],
        days=["%d %B %Y"],
        months=["%d %B %Y"],
        years=["%d %B %Y"],
    ))
p2.xaxis.major_label_orientation = pi/4
show(p2)
```

```python
#df_grades_indexed = df_grades.set_index('grade_date')
df_grades_indexed=df_grades2.pivot(index='grade_date', columns='subject_name', values='grade')
df_grades_indexed
```

```python
import numpy as np
from bokeh.charts import TimeSeries
from bokeh.models import HoverTool
from bokeh.palettes import Spectral11

numlines=len(df_grades_indexed.columns)
mypalette=Spectral11[0:numlines]

_tools_to_show = 'box_zoom,pan,save,hover,resize,reset,tap,wheel_zoom' 

p5 = figure(width=500, height=300, x_axis_type="datetime", tools=_tools_to_show) 
p5.multi_line(xs=[df_grades_indexed.index.values]*numlines,
                ys=[df_grades_indexed[subject].values for subject in df_grades_indexed],
                line_color=mypalette,
                line_width=5)

# THEN put  scatter one at a time on top of each one to get tool tips (HACK! lines with tooltips not yet supported by Bokeh?) 
for (name, series) in df_grades_indexed.iteritems():
    # need to repmat the name to be same dimension as index
    name_for_display = np.tile(name, [len(df_grades_indexed.index),1])

    source = ColumnDataSource({'x': df_grades_indexed.index, 'y': series.values, 
                               'series_name': name_for_display, 'Date': df_grades_indexed.index.format()})
    # trouble formating x as datestring, so pre-formating and using an extra column. It's not pretty but it works.

    p5.scatter('x', 'y', source = source, fill_alpha=0, line_alpha=0.3, line_color="grey")     

    hover = p5.select(dict(type=HoverTool))
    hover.tooltips = [("Subject", "@series_name"), ("Date", "@Date"),  ("FinalAvg", "@y"),]
    hover.mode = 'mouse'


output_notebook()
show(p5)
```
